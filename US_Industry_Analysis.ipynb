This project aims to determine which towns are dominated by each of the four sectors of industry.

```python
import pandas as pd
```

The source of my data is a CSV provided by a college professor. This was scraped from www.uszip.com

```python
zipcode_df = pd.read_csv("http://www1.assumption.edu/users/alfano/CSC-233-SP19/USZIP00000to99999_StateFound.csv")
```

```python
zipcode_df.columns
```

To start, I created a dataframe filled with locations and specific industries.

```python
industry_df = zipcode_df[['zip_code', 'town', 'state', 'agriculture_/_forestry_/_fishing_/_hunting_/_mining', 'construction', 'manufacturing', 'wholesale_trade', 'retail_trade', 'transportation_/_warehousing_/_utilities', 'information', 'finance_/_insurance_/_real_estate_/_rental_/_leasing', 'professional_/_scientific_/_management_/_administrative_/_waste_management_services', 'educational_services_/_health_care_/_social_assistance', 'arts_/_entertainment_/_recreation_/_accommodation_/_food_services', 'public_administration', 'other_services', 'civilian_employed_population_(16_years_and_over)']]
```

```python
industry_df.head(10)
```

I created a dataframe similar to the previous, but each industry is in terms of percent for the purpose of observing by proportion.

```python
industry_percent_df = industry_df[['agriculture_/_forestry_/_fishing_/_hunting_/_mining', 'construction', 'manufacturing', 'wholesale_trade', 'retail_trade', 'transportation_/_warehousing_/_utilities', 'information', 'finance_/_insurance_/_real_estate_/_rental_/_leasing', 'professional_/_scientific_/_management_/_administrative_/_waste_management_services', 'educational_services_/_health_care_/_social_assistance', 'arts_/_entertainment_/_recreation_/_accommodation_/_food_services', 'public_administration', 'other_services', 'civilian_employed_population_(16_years_and_over)']]
industry_percent_df = industry_percent_df.apply(lambda x: (x/industry_percent_df['civilian_employed_population_(16_years_and_over)'])*100)
industry_percent_df = industry_percent_df.drop('civilian_employed_population_(16_years_and_over)', axis=1)

industry_percent_df
```

```python
industry_percent_df['zip_code'] = industry_df[['zip_code']]
industry_percent_df['town'] = industry_df[['town']]
industry_percent_df['state'] = industry_df[['state']]
industry_percent_df = industry_percent_df[['state', 'town', 'zip_code', 'agriculture_/_forestry_/_fishing_/_hunting_/_mining', 'construction', 'manufacturing', 'wholesale_trade', 'retail_trade', 'transportation_/_warehousing_/_utilities', 'information', 'finance_/_insurance_/_real_estate_/_rental_/_leasing', 'professional_/_scientific_/_management_/_administrative_/_waste_management_services', 'educational_services_/_health_care_/_social_assistance', 'arts_/_entertainment_/_recreation_/_accommodation_/_food_services', 'public_administration', 'other_services']]


industry_percent_df.head(10)
```

As a precautionary measure, I removed any rows containing NaN values for industry.

```python
industry_percent_df = industry_percent_df[industry_percent_df['agriculture_/_forestry_/_fishing_/_hunting_/_mining'].notnull()]
```

```python
industry_percent_df.columns
```

This dataframe can now be used to query and organize the data by the following sectors:<BR>
<UL>
    <LI>Primary (raw materials): agriculture/forestry/fishing/hunting/mining</LI>
    <LI>Secondary: construction, manufacturing</LI>
    <LI>Tertiary (service): wholesale trade, retail trade, transportation/warehousing/utilities, finance/insurance/real estate/rental/leasing, professional/scientific/management/administrative/waste management services, educational services/health care/social assistance, arts/entertainment/recreation/accommodation/food services, public administration, other services</LI>
    <LI>Quaternary (intellectual services): information</LI>
    </UL>

I constructed a new dataframe that combines industries into their respective sectors.

```python
industry_sectors_df = industry_percent_df[['zip_code', 'town', 'state']]
industry_sectors_df.loc[:,'Primary'] = industry_percent_df.loc[:,'agriculture_/_forestry_/_fishing_/_hunting_/_mining']
industry_sectors_df.loc[:,'Secondary'] = industry_percent_df.loc[:,'construction'] + industry_percent_df.loc[:,'manufacturing']
industry_sectors_df.loc[:,'Tertiary'] = industry_percent_df.loc[:,'wholesale_trade'] + industry_percent_df.loc[:,'retail_trade'] + industry_percent_df.loc[:,'transportation_/_warehousing_/_utilities'] + industry_percent_df.loc[:,'finance_/_insurance_/_real_estate_/_rental_/_leasing'] + industry_percent_df.loc[:,'professional_/_scientific_/_management_/_administrative_/_waste_management_services'] + industry_percent_df.loc[:,'educational_services_/_health_care_/_social_assistance'] + industry_percent_df.loc[:,'arts_/_entertainment_/_recreation_/_accommodation_/_food_services'] + industry_percent_df.loc[:,'public_administration'] + industry_percent_df.loc[:,'other_services']
industry_sectors_df.loc[:,'Quaternary'] = industry_percent_df.loc[:,'information']

industry_sectors_df
```

In order to determine which towns are dominated by each sector, I queried the data in terms of sectors that are at least 75 percent of the total industry. I left the data sorted by location as it was before in case of the presence of geographic sector hubs. These hubs would be states or clusters of towns that are generally dominated by the same sector. I also listed unique values of states with frequency to get an idea of where certain sectors are dominant.

```python
industry_sectors_df.query("Primary >= 75")
```

Quaternary sector:

```python
industry_sectors_df.query("Quaternary >= 75")
```

These results show that the tertiary sector is the most prevalent. This is understandable considering how widespread and various its subindustries are. The primary and secondary sectors are more scattered. These sectors are substantial and necessary for other sectors, but only require fewer distinct locations. The quaternary sector is the most scarce of the four sectors. This is because information is a relatively new and rising industry. I expect that the quaternary sector will continue to grow and eventually be on par with the other sectors.
